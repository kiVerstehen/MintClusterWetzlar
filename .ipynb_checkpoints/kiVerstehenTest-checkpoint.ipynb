```python
import random
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from ipywidgets import interactive, widgets
from IPython.display import display, clear_output
import numpy as np

def hundOderKatzeAnhandGewicht():
    
    font_path = "/usr/share/fonts/truetype/ancient-scripts/Symbola.ttf"  # anpassen!
    emoji_font = fm.FontProperties(fname=font_path)
    
    # Daten
    cat_heights = [20, 24, 31, 44, 50]
    cat_weights = [30, 20, 15, 25, 30]

    dog_heights = [18, 30, 35, 60, 45]
    dog_weights = [37, 27, 35, 30, 38]

    def plot_counting(y_achsenabschnitt=0.0):
        clear_output(wait=True)
        fig, ax = plt.subplots()

        # Entscheidungsgrenze
        x_vals = np.linspace(10, 70, 100)
        y_vals = y_achsenabschnitt + 0 * x_vals
        ax.plot(x_vals, y_vals, color='royalblue', label=f'Grenze bei {y_achsenabschnitt:.2f} kg')

        # Bereich einfärben
        ax.fill_between(x_vals, y_vals, 12, color='lightblue', alpha=1, label='kategorisiert als Katze')
        ax.fill_between(x_vals, y_vals, 42, where=(y_vals < 42), color='navajowhite', alpha=1, label='kategorisiert als Hund')

  

        # Katzen einzeichnen
        for i in range(len(cat_heights)):
            if cat_weights[i] > y_achsenabschnitt:
                emoji = '😿'  # falsch
            else:
                emoji = '😸'  # korrekt
            ax.text(cat_heights[i], cat_weights[i], emoji,fontproperties=emoji_font, fontsize=35, ha='center')
        
        # Hunde einzeichnen
        for i in range(len(dog_heights)):
            if dog_weights[i] < y_achsenabschnitt:
                emoji = '🐺'  # falsch
            else:
                emoji = '🐶'  # korrekt
            ax.text(dog_heights[i], dog_weights[i], emoji, fontsize=35, ha='center')

        ax.set_xlim(10, 68)
        ax.set_ylim(12, 42)
        ax.set_xlabel('Größe (cm)')
        ax.set_ylabel('Gewicht (kg)')
        ax.legend()

        plt.show()

    randomY = random.uniform(10, 50)
    w = interactive(plot_counting, y_achsenabschnitt=widgets.FloatSlider(min=10, max=50, step=0.05, value=randomY))
    display(w)

hundOderKatzeAnhandGewicht()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os

# Beispiel-Emoji-String
emojis = "🐱😸😿🐶🐺"

# Alle installierten Fonts abrufen
font_paths = fm.findSystemFonts(fontpaths=None, fontext='ttf')

# Anzahl der Vorschauen beschränken (optional)
max_fonts_to_test = None  # kannst du erhöhen oder auf None setzen

# Vorschau der Fonts mit Emoji-Test
for i, font_path in enumerate(font_paths):
    if max_fonts_to_test and i >= max_fonts_to_test:
        break
    try:
        font = fm.FontProperties(fname=font_path)
        font_name = font.get_name()

        fig, ax = plt.subplots()
        ax.text(0.5, 0.5, emojis, fontproperties=font, fontsize=30, ha='center', va='center')
        ax.set_title(f"Font: {font_name}", fontsize=12)
        ax.axis("off")
        plt.show()
    except Exception as e:
        print(f"Fehler mit {font_path}: {e}")
```
